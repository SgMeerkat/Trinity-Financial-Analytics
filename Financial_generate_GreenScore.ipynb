```python
# Dictionary mapping engine types to highway and city fuel consumption rates
highway_fuel_consumption = {
    'F8D Petrol Engine': 6.75,
    '1.2 L K12N Dualjet': 5.85,
    '1.0 SCe': 4.95,
    '1.5 L U2 CRDi': 7.65,
    '1.5 Turbocharged Revotorq': 7.2,
    'K Series Dual jet': 5.4,
    '1.2 L K Series Engine': 5.85,
    'K10C': 4.5,
    'i-DTEC': 6.75,
    'G12B': 5.4,
    '1.5 Turbocharged Revotron': 7.2
}

city_fuel_consumption = {
    'F8D Petrol Engine': 9,
    '1.2 L K12N Dualjet': 7.8,
    '1.0 SCe': 6.6,
    '1.5 L U2 CRDi': 10.2,
    '1.5 Turbocharged Revotorq': 9.6,
    'K Series Dual jet': 7.2,
    '1.2 L K Series Engine': 7.8,
    'K10C': 6,
    'i-DTEC': 9,
    'G12B': 7.2,
    '1.5 Turbocharged Revotron': 9.6
}
# Applying a 2.5x multiplier to all consumption rates
highway_fuel_consumption = {key: value * 2.5 for key, value in highway_fuel_consumption.items()}
# Applying a 2.5x multiplier to all consumption rates
city_fuel_consumption = {key: value * 2.5 for key, value in city_fuel_consumption.items()}


# Create new columns for highway and city fuel consumption rates
df['fuel_consumption_Hwy (L/100 km)'] = df['engine_type'].map(highway_fuel_consumption)
df['fuel_consumption_City (L/100 km)'] = df['engine_type'].map(city_fuel_consumption)

# Display the DataFrame with the new columns
print(df.head())
```

```python
# Define age factor for each age bin
age_factor_bins = {
    (0, 0.05): 1,   # New cars, no age factor
    (0.05, 0.1): 1.2,  # Slightly aged cars
    (0.1, 0.15): 1.5,  # Moderately aged cars
    (0.15, 1): 2      # Old cars, higher age factor
}

# Function to apply age factor based on age_of_car
def apply_age_factor(age):
    for age_range, factor in age_factor_bins.items():
        if age_range[0] <= age < age_range[1]:
            return factor
    return 1  # Default factor for unknown ages

# Apply age factor to adjust fuel consumption rates
df['age_factor'] = df['age_of_car'].apply(apply_age_factor)
df['fuel_consumption_Hwy (L/100 km)'] *= df['age_factor']
df['fuel_consumption_City (L/100 km)'] *= df['age_factor']

# Drop age_factor column
df.drop('age_factor', axis=1, inplace=True)

# Display the DataFrame with the adjusted fuel consumption columns
print(df.head())
```
